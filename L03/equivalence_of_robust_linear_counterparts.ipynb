```python
import rsome as rso
import numpy as np
from rsome import ro
from rsome import msk_solver as my_solver  #Import Mosek solver interface
#from rsome import grb_solver as my_solver  #Import Gurobi solver interface
```

```python
# Parameters setup
n  = 150
i = np.arange(1, n+1)
c  = 0.15+ i*0.05/150
a = np.zeros(n)
b = 0.02

# Describing the uncertainty
maxDev = 0.05/450 * (2*i*n*(n+1))**0.5
uBarPlus = maxDev
uBarMinus = -maxDev
```

```python
import matplotlib.pyplot as plt
plt.plot(uBarPlus)
plt.plot(uBarMinus)
plt.xlabel('i')
plt.ylabel('u+ and u-')
plt.grid()
plt.show()
```

## **Solving Deterministic Model**

Consider the following linear programming problem:
\begin{align}
\max\limits_{x}\;\;&c^\top x \\
\text{subject to}\;\;& a^\top x \leq b\\
& 0 \leq x \leq 1.
\end{align}

```python
# Create model
model = ro.Model('simpleExample_det')# Create a optimization model using rsome

# Define the decision variable vector x of dimension n
x=model.dvar(n) # By default, these are continuous optimization variables

# Set the objective function:
model.max(c@x) # maximize c^T x

# Add the main linear constraint:
model.st(a@x<=b) # a^T x <= b
model.st(x>=0) # x_i >= 0 for all i
model.st(x<=1) # x_i <= 1 for all i

# Solve the optimization problem with the chosen solver
model.solve(my_solver)
optobj_det = model.get() # Retrieve the optimal objective value
xx_det   = x.get() # Retrieve the optimal values of the decision variables

# Print results
print('objective c^Tx=')
print(optobj_det)
print('solution x=')
print(xx_det)
```

## **Solving Raw Robust Counterpart**

Consider the following robust linear optimization problem:
\begin{align}
\max\limits_{x}\;\;&c^\top x \\
\text{subject to}\\ \;\;& (a + u)^\top x \leq b && \forall u \in \mathcal{U}\\
& 0 \leq x \leq 1,
\end{align}
where $$\mathcal{U}:=\{u\;|\;\bar{u}^- \leq u \leq \bar{u}^+\}.$$

```python
#Create model
model = ro.Model('simpleExample_rawrobust')
x=model.dvar(n)

#Create uncertain vector
u = model.rvar(n)

#Create uncertainty set
USet = (u>=uBarMinus, u <= uBarPlus)
model.max(c@x)

#Robustify the constraint
model.st(((a+u)@x<=b).forall(USet))  # apply robustness using .forall()
model.st(x>=0)
model.st(x<=1)

# Solve the optimization problem with the chosen solver
model.solve(my_solver)

optobj_raw = model.get() # Retrieve the optimal objective value
xx_raw   = x.get() # Retrieve the optimal values of the decision variables

print('objective',optobj_raw)
print('solution',xx_raw)
```

## **Solving Reduced Robust Counterpart**

The robust counterpart can be reformulated as follows:

\begin{align}
\max\limits_{x,\lambda^+,\lambda^-}\;\;&c^\top x \\
\text{subject to}\\
\;\;& a ^\top x + (\bar{u}^+)^\top \lambda^+ - (\bar{u}^-)^\top \lambda^- \leq b, &&\\
& \lambda^+ -\lambda^- =x, \\
& \lambda^- \geq 0, \\
& \lambda^+ \geq 0,\\
& 0 \leq x \leq 1.
\end{align}

```python
# Create the robust optimization model
model = ro.Model('simpleExample_redrobust')

# Define the decision variable vector x of dimension n
x = model.dvar(n)

# Define the auxiliary dual variables associated with
# the upper and lower bounds of the box uncertainty set
lambdaPlus = model.dvar(n)
lambdaMinus = model.dvar(n)

# Set the objective function:
# maximize c^T x
model.max(c @ x)

# Add the robust counterpart of the uncertain constraint.
# This comes from dualizing the inner worst-case problem
# over the box uncertainty set.
model.st(a @ x + uBarPlus @ lambdaPlus - uBarMinus @ lambdaMinus <= b)

# Add the equality constraint from the dual formulation:
# lambdaPlus - lambdaMinus = x
model.st(lambdaPlus - lambdaMinus == x)

# Enforce nonnegativity of the dual variables
model.st(lambdaPlus >= 0)
model.st(lambdaMinus >= 0)

# Add bounds on the decision variables:
# 0 <= x <= 1
model.st(x >= 0)
model.st(x <= 1)

# Solve the optimization problem with the chosen solver
model.solve(my_solver)

# Retrieve the optimal objective value
optobj_red = model.get()

# Retrieve the optimal solution for x
xx_red = x.get()

# Print results
print('objective', optobj_red)
print('solution', xx_red)
```
